```python
def extract(full):
    def _ex(a, n):
        return list(map(lambda x: x[n], a))
    
    def _abs(a):
        return list(map(lambda x: abs(x), a))
    
    return _ex(full, 0), _ex(full, 1), _abs(_ex(full, 2)), _abs(_ex(full, 3))
```

```python
def plot_with_errors(x, y, x_err, y_err, title=None, x_spacing=0.8, y_spacing=0.05):
    fig, ax = plt.subplots()
    ax.set_title(title, fontsize=20, fontweight='bold')
    
    ax.hlines(y=0, xmin=min(x), xmax=max(x), color='grey', label='0')
    ax.vlines(x=0, ymin=min(y), ymax=max(y), color='grey', label='0')
    
    ax.xaxis.set_ticks(np.arange(min(x), max(x), x_spacing))
    ax.yaxis.set_ticks(np.arange(min(y), max(y), y_spacing))
    
    ax.set_xlabel("V [V]", loc='right')
    ax.set_ylabel("I [A]", loc='top')
        
    ax.grid()
    
    ax.errorbar(x, y, xerr=x_err, yerr=y_err, fmt="o-", ecolor="red", barsabove=True)
```

```python
plot_with_errors(*extract(lamp), title="Lampara", x_spacing=0.8, y_spacing=0.05)
```

```python
plot_with_errors(*extract(resistor), title="Resistencia", x_spacing=0.8, y_spacing=0.05)
```

```python
plot_with_errors(*extract(diode), title="Diodo", x_spacing=0.08, y_spacing=0.05)
```
